```
def ShowPrice(dfs, labels, price='Open'):
  f  = plt.figure(figsize=[10,7])
  ax = f.add_subplot(1,1,1)

  for i in range(len(dfs)):
    ax.plot(dfs[i][price], label=labels[i])
  
  ax.legend(loc='best')
  ax.grid(True)
 
```

```
def ShowRelativePrice(dfs, labels, price='Open'):
  f  = plt.figure(figsize=[10,7])
  ax = f.add_subplot(1,1,1)

  for i in range(len(dfs)):
    ax.plot(dfs[i][price]/dfs[i][price].min(), label=labels[i])
  
  ax.legend(loc='best')
  ax.grid(True)
 
```

```
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_scaled_dict = {}

for file_name in files:
  print(file_name)
  # 标准化输入数据
  scaled_data = scaler.fit_transform(df_dict[file_name].drop('Unnamed: 0', 1).values)
  df_scaled_dict[file_name] = pd.DataFrame(scaled_data, index=df_dict[file_name].drop('Unnamed: 0', 1).index, columns=df_dict[file_name].drop('Unnamed: 0', 1).columns)

  # 填充缺失数据
  df_scaled_dict[file_name].fillna(method='ffill', inplace=True) # 按前一天数据填充缺失值
  df_scaled_dict[file_name].fillna(0, inplace=True) # 如果数据从一开始就缺失，填0

  print(df_scaled_dict[file_name].head(n=2))
```

```
def getXTrain(df, enddate, periods):
  dates =  pd.date_range(end=enddate, periods=periods)
  out_list = df.reindex(dates).fillna(method='ffill').fillna(0).values

  return out_list
```

```
import keras
from keras.layers import Dense, Dropout, LSTM, Input

#======================================================================================================        
class LossHistory(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.losses = []

    def on_batch_end(self, batch, logs={}):
        self.losses.append(logs.get('loss'))

#======================================================================================================        
def trainRNN(comex_list, LME_list, indices, IO, Y_labels):
  #
  # comex RNN model
  #
  comex_periods  = comex_list[0].shape[1]
  comex_features = comex_list[0].shape[2]
  
  input_comex  = keras.layers.Input(shape=(comex_periods, comex_features))
  masked_comex = keras.layers.Masking()(input_comex)
  lstm_comex   = LSTM(100, return_sequences=False)(masked_comex)
  comex_model  = keras.models.Model(inputs=input_comex, outputs=lstm_comex)

  # feed inputs
  lstm_comex_outs = []
  lstm_comex_inps = []
   
  for i in range(len(comex_list)):
    lstm_comex_inps += [Input(shape=(comex_periods, comex_features))]
    lstm_comex_outs += [comex_model(lstm_comex_inps[-1])]        

  merged_comex    = keras.layers.concatenate(lstm_comex_outs)
  comex_dpt1      = keras.layers.Dropout(0.2)(merged_comex)
  comex_dense1    = Dense(100, activation='relu')(comex_dpt1)

  # 
  # lme RNN model
  #
  lme_periods  = LME_list[0].shape[1]
  lme_features = LME_list[0].shape[2]

  input_lme   = keras.layers.Input(shape=(lme_periods, lme_features))
  masked_lme  = keras.layers.Masking()(input_lme)
  lstm_lme    = LSTM(100, return_sequences=False)(masked_lme)
  lme_model   = keras.models.Model(inputs=input_lme, outputs=lstm_lme)

  # feed inputs
  lstm_lme_outs = []
  lstm_lme_inps = []
   
  for i in range(len(LME_list)):
    lstm_lme_inps += [Input(shape=(lme_periods, lme_features))]
    lstm_lme_outs += [lme_model(lstm_lme_inps[-1])]        

  merged_lme    = keras.layers.concatenate(lstm_lme_outs)
  lme_dpt1      = keras.layers.Dropout(0.2)(merged_lme)
  lme_dense1    = Dense(100, activation='relu')(lme_dpt1)
  
  #
  # Other models
  # 
  input_indices  = keras.layers.Input(shape=(indices.shape[1], indices.shape[2]))
  masked_indices = keras.layers.Masking()(input_indices)
  lstm_indices   = LSTM(40, return_sequences=False)(masked_indices)

  input_IO  = keras.layers.Input(shape=(IO.shape[1], IO.shape[2]))
  masked_IO = keras.layers.Masking()(input_IO)
  lstm_IO   = LSTM(40, return_sequences=False)(masked_IO)

  # Dropout and merge
  merged_all = keras.layers.concatenate([comex_dense1, lme_dense1, lstm_indices, lstm_IO])
  
  dpt1 = keras.layers.Dropout(0.3)(merged_all)
  fc1 = keras.layers.Dense(40, activation='relu')(dpt1)
  dpt2 = keras.layers.Dropout(0.3)(fc1)
  fc2 = keras.layers.Dense(40, activation='relu')(dpt2)

  output = keras.layers.Dense(6, activation='sigmoid', name="signal")(fc2)

  model_inp = lstm_comex_inps + lstm_lme_inps + [input_indices, input_IO]
  model = keras.models.Model(inputs=model_inp, outputs=output)
  model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])

  print(model.summary())
  history = LossHistory()

  fhist = model.fit(comex_list + LME_list + [indices, IO],
                    Y_labels,
                    validation_split=0.1,
                    epochs=30,
                    #validation_data=(test_sets, test_labels),
                    batch_size=512,
                    callbacks=[history])
  
  keras.utils.plot_model(model, show_shapes=True)
  return model
    
```

```
dict_names = [
           "COMEX_Copper_train", 
           "COMEX_Gold_train", 
           "COMEX_Palladium_train", 
           "COMEX_Platinum_train", 
           "COMEX_Silver_train", 
           "LMEAluminium3M_train", 
           "LMECopper3M_train", 
           "LMELead3M_train", 
           "LMENickel3M_train", 
           "LMETin3M_train", 
           "LMEZinc3M_train", 
           "Indices_all",
           "IO_all"]

X_train_dict = {}

for dict_name in dict_names:
  x_train_date_list = []

  for enddate in Y_label_1d_df.index.values:
    x_train_date_list += [getXTrain(df_scaled_dict[dict_name], enddate, 28)]
  
  X_train_dict[dict_name] = np.array(x_train_date_list)
```

```
comex = ["COMEX_Copper_train", 
         "COMEX_Gold_train", 
         "COMEX_Palladium_train", 
         "COMEX_Platinum_train", 
         "COMEX_Silver_train"]

comex_list = [X_train_dict[dict_name] for dict_name in comex]

lme   = [
           "LMEAluminium3M_train", 
           "LMECopper3M_train", 
           "LMELead3M_train", 
           "LMENickel3M_train", 
           "LMETin3M_train", 
           "LMEZinc3M_train", ]

lme_list = [X_train_dict[dict_name] for dict_name in lme]

model = trainRNN(comex_list, lme_list, X_train_dict["Indices_all"], X_train_dict["IO_all"],  Y_label_1d_df.values)
```
